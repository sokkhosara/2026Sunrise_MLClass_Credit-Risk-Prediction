```python
# Flask server
from flask import Flask, request, jsonify 
import pickle

# Load model once when app starts
with open('KM_model_CreditRisk=1.0.bin', 'rb') as f:
    dv, model = pickle.load(f)

app = Flask('predict')
@app.route('/predict', methods = ['POST'])
def predict():
     customer = request.get_json()

     # convert to dict format expected by dv
     dicts = [customer]

     # transform using SAME dv
     X = dv.transform(dicts)

     # predict probability
     y_pred = model.predict_proba(X)[0, 1]

     # convert to binary
     risk_flag = int(y_pred >= 0.5)
     
     return jsonify({
        "risk_probability": float(y_pred),
        "risk_flag": risk_flag
    })
     

if __name__ == "__main__":
    app.run(debug=True, host='0.0.0.0',port=9696, use_reloader=False)
```
